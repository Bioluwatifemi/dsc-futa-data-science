```python
x = [1,2,3,4,5]
y = [4,4,3,4,4]

plt.plot(x, y, label = '1st line')
plt.plot(y, x, label = '2nd line')
plt.plot(x, np.sin(x), label = 'curve 1')
plt.plot(x, np.cos(x), label = 'curve 2')
plt.legend()
plt.show()
```
